```python
from skimage.io import imread
import matplotlib.pyplot as plt
from cellpose import models, io

# Load the image
image = imread('blobs.tif')

# Initialize Cellpose model
model = models.Cellpose(model_type='cyto')

# Segment the bright blobs using Cellpose
masks, flows, styles, diams = model.eval(image, diameter=None, channels=[0,0])

# Display the original image and the segmented labels
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow(image, cmap='gray')
ax[0].set_title('Original Image')
ax[0].axis('off')

ax[1].imshow(masks, cmap='jet')
ax[1].set_title('Segmented Labels')
ax[1].axis('off')

plt.show()
```
