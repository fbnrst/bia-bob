```python
import numpy as np
from scipy.ndimage import zoom
from aicsimageio import AICSImage
import stackview

# Load the image file
filename = "Image_6_Subset_pos02_t1_35.czi"
aics_image = AICSImage(filename)

# Retrieve the image data and metadata
image_data = aics_image.get_image_data("CZYX")
metadata = aics_image.metadata

# Extract the pixel size information for x, y, and z dimensions
pixel_size_x = metadata.images[0].pixels.physical_size_x
pixel_size_y = metadata.images[0].pixels.physical_size_y
pixel_size_z = metadata.images[0].pixels.physical_size_z

# Choose a target pixel size (e.g., the smallest pixel size among x, y, z)
target_pixel_size = min(pixel_size_x, pixel_size_y, pixel_size_z)

# Compute the scaling factors
scaling_factors = [
    pixel_size_z / target_pixel_size,
    pixel_size_y / target_pixel_size,
    pixel_size_x / target_pixel_size
]

# Rescale the image data
rescaled_image = np.zeros((
     image_data.shape[0], 
    int(image_data.shape[1] * scaling_factors[0]),
    int(image_data.shape[2] * scaling_factors[1]),
    int(image_data.shape[3] * scaling_factors[2]))
)
for channel in range(image_data.shape[0]):
    rescaled_image[channel] = zoom(image_data[channel], scaling_factors, order=1)  # Using linear interpolation

# Ensure the data shape is correct
rescaled_image = rescaled_image.astype(np.float32)

# Display the rescaled image
stackview.slice(rescaled_image)
```
